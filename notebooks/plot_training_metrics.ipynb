```python
import pandas as pd
import plotly.graph_objects as go
metrics_ltc = pd.read_csv("gesture_ltc.csv")
metrics_cfc = pd.read_csv("gesture_cfc.csv")
df_val_metrics_ltc = metrics_ltc.drop("train_loss", axis=1).dropna()
df_val_metrics_cfc = metrics_cfc.drop("train_loss", axis=1).dropna()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_val_metrics_ltc["epoch"], y=df_val_metrics_ltc["val_accuracy"], mode='lines', name='LTC'))
fig.add_trace(go.Scatter(x=df_val_metrics_cfc["epoch"], y=df_val_metrics_cfc["val_accuracy"], mode='lines', name='CfC'))
fig.add_trace(go.Scatter(x=[0,200], y=[0.6955, 0.6955], mode='lines', name='Best result of the article', line=dict(dash="dash")))
fig.update_layout(title="Gesture Phase Segmentation: validation accuracy", template="simple_white")
fig.update_yaxes(title="Accuracy", showgrid=True, range = [0,1] , dtick=0.1)
fig.update_xaxes(title="Epoch", showgrid=True, range=[0,201], dtick=20)
fig.update_layout(width=576, height=384)
# Move legend inside the plot
fig.update_layout(
    legend=dict(
        x=0.5,   # Horizontal position (closer to 1 moves it to the right)
        y=0.0,   # Vertical position (closer to 1 moves it to the top)
        bgcolor="rgba(255, 255, 255, 0.5)"  # Optional: semi-transparent background
    )
)

fig.show()
# Save the figure as a PDF

fig.write_image("Gesture_accuracy.pdf", format="pdf")
```

```python
import pandas as pd
import plotly.graph_objects as go
metrics_ltc = pd.read_csv("power_ltc.csv")
metrics_cfc = pd.read_csv("power_cfc.csv")
df_val_metrics_ltc = metrics_ltc.drop("train_loss", axis=1).dropna()
df_val_metrics_cfc = metrics_cfc.drop("train_loss", axis=1).dropna()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_val_metrics_ltc["epoch"], y=df_val_metrics_ltc["val_loss"], mode='lines', name='LTC'))
fig.add_trace(go.Scatter(x=df_val_metrics_cfc["epoch"], y=df_val_metrics_cfc["val_loss"], mode='lines', name='CfC'))
fig.add_trace(go.Scatter(x=[0,5], y=[0.586, 0.586], mode='lines', name='Best result of the article', line=dict(dash="dash")))
fig.update_layout(title="Individual Household Electric Power Consumption: <br>validation MSE", template="simple_white")
fig.update_yaxes(title="MSE", showgrid=True)
fig.update_xaxes(title="Epoch", showgrid=True, range=[0,10])
fig.update_layout(width=576, height=384)
# Set y-axis to log scale (base 10)
# Customize y-axis
fig.update_layout(
    yaxis=dict(
        type="log",      # Logarithmic scale (Base 10)
        dtick=1,         # Show only powers of 10 (10, 100, 1000)
        minor=dict(
            ticks="inside",  # Show minor ticks
            showgrid=True    # Add gridlines for each tenth increment
        )
    )
)
# Move legend inside the plot
fig.update_layout(
    legend=dict(
        x=0.5,   # Horizontal position (closer to 1 moves it to the right)
        y=0.8,   # Vertical position (closer to 1 moves it to the top)
        bgcolor="rgba(255, 255, 255, 0.5)"  # Optional: semi-transparent background
    )
)
fig.show()
fig.write_image("Power_MSE.pdf", format="pdf")
```
